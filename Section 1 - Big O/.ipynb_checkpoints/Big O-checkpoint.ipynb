**Big O: O(n)**<br/>
Implement a Python function called ```print_items```.<br/>
This function should take a single integer as an argument and print out a sequence of numbers from 0 up to, but not including, the provided integer.<br/>
The function should use a ```for``` loop and Python's built-in ```range``` function to generate the sequence of numbers.<br/>
The function signature should be: ```def print_items(n)```.

````python
def print_items(n):
    for i in range(n):
        print(i)

# print_items(5)
````

**Big O: O(n^2)**<br/>
In this task, you are required to implement a function called ```print_items(n)```.<br/>
This function takes one integer argument, ```n```, and prints out a pair of numbers ```(i, j)``` for every possible combination of ```i``` and ```j``` in the range of 0 up to but not including ```n```.<br/>
Each pair ```(i, j)``` is printed on a new line. ```i``` and ```j``` iterate over the range independently, so every combination of ```i``` and ```j``` in the range should be printed.<br/>
The output of the function should be a series of pairs, printed one after the other, each on a new line. The pairs should be printed in the order that they are generated by the nested loops.

````python
def print_items(n):
    for i in range(n):
        for j in range(n):
            print(i,j)

print_items(3)
````
